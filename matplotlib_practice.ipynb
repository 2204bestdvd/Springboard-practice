```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Histogram

```python
x = np.random.normal(size=1000)
fig, ax = plt.subplots()
H = ax.hist(x, bins=50, alpha=0.5, histtype='stepfilled')
```

# Pie plot

```python
fracs = [30, 15, 45, 10]
colors = ['b', 'g', 'r', 'w']
fig, ax = plt.subplots(figsize=(6, 6))
pie = ax.pie(fracs, colors=colors, explode=(0, 0.1, 0.05, 0), shadow=False,
             labels=['A', 'B', 'C', 'D'])
```

# Errorbar Plots

```python
x = np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)

fig, ax = plt.subplots()
plt.errorbar(x, y, dy, fmt='.k')
```

# Filled Plots


```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where=(y1<y2), color='blue')
ax.fill_between(x, y1, y2, where=(y1>y2), color='green')
```

# Scatter Plots

```python
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50)  # color of points
s = 500 * np.random.random(50)  # size of points

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap=plt.cm.jet)

# Add colorbar
fig.colorbar(im, ax=ax)

# set the color limits
im.set_clim(0.0, 1.0)
```

# Contour Plots

```python
x = np.linspace(0, 10, 50)
y = np.linspace(0, 20, 60)
z = np.cos(y[:, np.newaxis]) * np.sin(x)

fig, ax = plt.subplots()

# filled contours
im = ax.contourf(x, y, z, 100, cmap=plt.cm.jet)
# contour lines
im2 = ax.contour(x, y, z, colors='k')

fig.colorbar(im, ax=ax)
```

# Showing Images

```python
I = np.random.random((100, 100))
I += np.cos(np.linspace(0, np.pi, 100))
fig, ax = plt.subplots()
im = ax.imshow(I, cmap=plt.cm.jet)
fig.colorbar(im, ax=ax)
```

# 2D Histogram and Hexbins

```python
x, y = np.random.normal(size=(2, 1000))
fig, ax = plt.subplots()
im = ax.hexbin(x, y, gridsize=20, cmap=plt.cm.jet)
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
H = ax.hist2d(x, y, bins=20, cmap=plt.cm.jet)
fig.colorbar(H[3], ax=ax)
```

# Polar Plots

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='polar')

theta = np.linspace(0, 10*np.pi, 1000)
r = np.linspace(0, 10, 1000)

ax.plot(theta, r)
```
